```python
import pandas as pd
data = pd.read_excel("/content/Date-Wise-Prices-all-Commodity.xlsx")
```

```python
data
```

```python
# prompt: Using dataframe data: check if each line contains na if contains delete the row

# Drop rows with missing values
data = data.dropna()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Commodity'] = le.fit_transform(data['Commodity'])
data['Market'] = le.fit_transform(data['Market'])
data['Variety'] = le.fit_transform(data['Variety'])
data['Grade'] = le.fit_transform(data['Grade'])
```

```python
data
```

```python
y = data['Modal_x0020_Price']
x= data[['Market','Commodity','Variety','Grade']]
```

```python
y.head()
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
import sklearn.metrics as metrics
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
```

```python
metrics.r2_score(y_test, y_pred)*100
```
