```python
import requests
url = 'https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv'
res = requests.get(url, allow_redirects=True)
with open('carprices.csv','wb') as file:
    file.write(res.content)
```

```python
import pandas as pd
data = pd.read_csv('/content/carprices.csv')
```

```python
# prompt: get all uniqque from a column in data

data['Car Model'].unique()
```

```python
# prompt: get dummies as 0 1

dummies = pd.get_dummies(data['Car Model'])
dummies
```

```python

dummies = dummies.replace({False: 0, True: 1}) # Replace False with 0 and True with 1 in all columns
dummies
```

```python
data = pd.concat([data, dummies], axis='columns')
data
```

```python
data = data.drop(['Car Model','Mercedez Benz C class'], axis='columns')
data
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x= data.drop(['Sell Price($)'], axis='columns')
y= data['Sell Price($)']
```

```python
x
```

```python
y
```

```python
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
model.score(x, y)
```
