```python
import pandas as pd
data = pd.read_csv("/content/salary-dataset-simple-linear-regression.zip")
data.head()
```

```python
# prompt: Using dataframe data: scatter plot using matplotlib

import matplotlib.pyplot as plt
plt.scatter(data['YearsExperience'], data['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
data.describe()

```

```python
data.drop(columns="Unnamed: 0",inplace= True)
data.drop_duplicates()
```

```python
x= data["YearsExperience"]
y= data["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train.values.reshape(-1,1),y_train)
```

```python
model.score(x_test.values.reshape(-1,1),y_test)
```

```python
y_pred = model.predict(x_test.values.reshape(-1,1))
```

```python
model.score(x_test.values.reshape(-1,1),y_pred)
```

```python
import sklearn.metrics as metrics
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
```

```python
metrics.r2_score(y_test, y_pred)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,color="red")
```
